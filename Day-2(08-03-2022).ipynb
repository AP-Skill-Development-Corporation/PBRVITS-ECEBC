## Assignment Operators



```python
t = 9

# 5 to t and re-assign the resultant value same variable
t += 5 # t = t + 5
print(t)
```

```python
n = 12

n -= 10
n *= 6
print(n) # 12

n %= 5
n**2
print(n) # 2

n += 100
n %= 8
print(n) # 6

n//10
print(n) #6

n *= 5
n **= 2
n/=10
print(n) # 90.0
```
